```python
import camelot
import numpy as np
import pandas as pd
import os
from sqlalchemy import create_engine
```

```python
os.getcwd()
```

```python
tables = camelot.read_pdf('BOE-A-2018-17664.pdf', pages='4-608')
```

```python
tables
```

```python
tables[-1].df.tail()
```

```python
df = pd.DataFrame()

for i in range(len(tables)):
    df_table = tables[i].df.iloc[2:]
    df = df.append(df_table)
```

```python
df.tail()
```

```python
# extract first column
df_manufacturers = df.iloc[:,0]

# remove duplicated manufacturers
df_manufacturers = df_manufacturers.drop_duplicates()

# remove empty fields
df_manufacturers.replace('', np.nan, inplace=True)
df_manufacturers.dropna(inplace=True)

df_manufacturers
```

```python
engine = create_engine('sqlite:///manufacturers.db')
df_manufacturers.to_sql('manufacturer', engine)
```

```python
columns = [
    'manufacturer',
    'model',
    'sales term',
    'cc',
    'cylinders',
    'fuel',
    'kW',
    'cvf',
    'HP',
    'value €'
]
```

```python
# Insert header

df.columns = columns

# df.loc[0] = columns
# df.index = df.index + 1
# df = df.sort_index()
```
